```python
import numpy as np
import matplotlib.pyplot as plt
#notebook指令，让matplotlib可以在notebook中显示绘图
%matplotlib inline 
```

```python
x=np.linspace(0,10,200)
y=np.sin(x)
plt.plot(x,y)
plt.plot(x,np.cos(x))
```

```python
#创建个画布再画图
fig =plt.figure()
plt.plot(x,y,'--')
#可以使用fig.savefig('路径')保存画布上得图片
```

```python
#多张图排版
x=np.linspace(0,10,80)
plt.subplot(2,1,1)#2行1列第1张图
plt.plot(x,np.sin(x),'o',color='red',label='sin(x)')
plt.plot(x,np.cos(x),label='cos(x)')
plt.legend()#有这句才能显示label
plt.subplot(2,1,2)#2行1列第2张图
plt.plot(x,np.cos(x),label='cos(x)')
plt.legend(loc=1)#loc 从0-10代表标签放置不同位置
```

```python
#散点图
x=np.linspace(0,10,20)
plt.scatter(x,np.sin(x))
```

```python
#散点图
x=np.random.randn(100)
y=np.random.randn(100)
colors=np.random.rand(100)
sizes=100*np.random.rand(100)
plt.scatter(x,y,c=colors,s=sizes)
plt.colorbar()
```
